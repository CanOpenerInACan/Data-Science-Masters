```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
obama = pd.read_excel('obama-approval-ratings.xls')
hotdog_winners = pd.read_excel('hotdog-contest-winners.xlsm')
hotdog_counts = pd.read_excel('hotdog-places.xlsm')
```

```python
plt.figure(figsize=(8,6))
plt.bar(hotdog_winners.Year, hotdog_winners['Dogs eaten'])
plt.xlabel('Year', fontsize=12), plt.ylabel('Hotdogs Eaten', fontsize=12)
plt.title('Hotdogs Eaten by Winner per Year\n', fontsize=14)
plt.grid(axis='y', linewidth='0.5')
sns.despine(top=True, right=True, left=True, bottom=False)
plt.show()
```

```python
hotdog_counts = hotdog_counts.T
hotdog_counts.rename(columns = {0:'First', 1:'Second', 2:'Third'}, inplace=True)
```

```python
firstarray = np.asarray(hotdog_counts.First)
secondarray = np.asarray(hotdog_counts.Second)


plt.figure(figsize=(8,6))
plt.bar(hotdog_counts.index, hotdog_counts.First)
plt.bar(hotdog_counts.index, hotdog_counts.Second, bottom=firstarray)
plt.bar(hotdog_counts.index, hotdog_counts.Third, bottom=firstarray+secondarray, color='red')
plt.xlabel('Year'), plt.ylabel('Hotdogs Eaten')
plt.title('Medalist Hotdog Counts by Year')
plt.legend(('First Place', 'Second Place', "Thid Place"))
plt.grid(axis='y')
sns.despine(top=True, right=True, left=True, bottom=False)
plt.show()
```

```python
counts = hotdog_winners.groupby('Country').agg('count').sort_values(by=['Winner'])
countries=['Germany', 'Japan', 'Mexico', 'United States']

plt.figure(figsize=(8,6))
plt.pie(counts.Winner, labels=countries, autopct='%1.1f%%', startangle=90, rotatelabels=True)
plt.title('Pie Chart of Winnner by Country', y=-0.01, fontsize=16)
plt.show()
```

```python
size_of_groups=[obama.Approve[0], obama.Disapprove[0], obama['None'][0]]
colors=["#F2BC94","#30110D", "#722620"]
labels=['Approve', 'Disapprove', 'No Opinion']


plt.figure(figsize=(8,6))
plt.pie(size_of_groups, colors=colors, labels=labels)
plt.title('Donut Chart of Obama Race Relations Opinions\n', fontsize=14)


my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```
