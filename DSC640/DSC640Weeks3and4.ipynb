```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pop = pd.read_excel('world-population.xlsm')
```

```python
fig, ax = plt.subplots()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(True)

plt.plot(pop.Year, pop.Population, color='k', linewidth=2.5)
ax.set_ylim(ymin=0, ymax=7000000000)
plt.title('Population (1960-2009) Line Chart\n', fontsize=14)
plt.xlabel('Year', fontsize=12), plt.ylabel('Population', fontsize=12)
plt.grid(axis='y', color='#D3D3D3')


plt.show()
```

```python
fig, ax = plt.subplots()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(True)

plt.step(pop.Year, pop.Population, color='k', linewidth=2)
ax.set_ylim(ymin=0, ymax=7000000000)
plt.title('Population (1960-2009) Step Chart\n', fontsize=14)
plt.xlabel('Year', fontsize=12), plt.ylabel('Population', fontsize=12)
plt.grid(axis='y', color='#D3D3D3')
plt.show()
```
