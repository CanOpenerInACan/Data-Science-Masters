# English Premier League Cutoffs (2000-2020)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
year = np.arange(2000,2021)
first = [91,80,87,83,90,95,91,89,87,90,86,80,89,89,86,87,81,93,100,98,99]
fourth = [67,68,71,67,60,61,67,68,76,72,70,68,69,73,79,70,66,76,75,71,66]
safety = [36,42,40,44,39,34,38,38,36,35,35,40,37,39,36,38,39,40,36,36,35]
```

```python
df = pd.DataFrame(list(zip(first, fourth, safety)), index=year, columns = ['1st Points', '4th Points', 'Safety'])
```

```python
df
```

```python
plt.figure(figsize=(13,8))
plt.plot(df['1st Points'], color='#132257')
plt.plot(df['4th Points'], color='#AA0000')
plt.plot(df['Safety'], color='#B3995D')
plt.title('Premier League Points Comparison by Year', fontsize=20)
plt.xlabel('Year', fontsize=17)
plt.ylabel('Points', fontsize=17)
plt.xticks([i for i in year])
plt.grid(True)
plt.legend(['Points of League Winner', 'Points of 4th Place Finisher', 'Points of Safety'])
```

```python
plt.figure(figsize=(10,6))

#First Place Points
plt.subplot(131)
plt.boxplot(df['1st Points'], showmeans=True, meanline=True, meanprops=dict(linestyle='--', color='#132257'))
plt.title('1st Place Points', fontsize=16)
plt.ylabel('Points', fontsize=12)
plt.yticks([i for i in range(80,101,2)], fontsize=12)

#Fourth Place Points
plt.subplot(132)
plt.boxplot(df['4th Points'], showmeans=True, meanline=True, meanprops=dict(linestyle='--', color='#AA0000'))
plt.title('4th Place Points', fontsize=16)
plt.ylabel('Points', fontsize=12)
plt.yticks([i for i in range(60,81, 2)], fontsize=12)

#Safety Points
plt.subplot(133)
plt.boxplot(df['Safety'], showmeans=True, meanline=True, meanprops=dict(linestyle='--', color='#B3995D'))
plt.title('Safety', fontsize=16)
plt.ylabel('Points', fontsize=12)
plt.yticks(fontsize=12)
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.09, right=0.95, hspace=0.5,
                    wspace=0.60)
```

```python
plt.figure(figsize=(10,6))

plt.subplot(131)
plt.hist(df['1st Points'], edgecolor='k', color='#132257', bins=int(df['1st Points'].max()-df['1st Points'].min()), align='left')
plt.xlabel('Point Count', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('1st Place Points', fontsize=16)
plt.xticks([i for i in range(80,100,2)], fontsize=10)
plt.yticks([i for i in range(0,4)], fontsize=12)

plt.subplot(132)
plt.hist(df['4th Points'], edgecolor='k', color='#AA0000', bins=int(df['4th Points'].max()-df['4th Points'].min()), align='left')
plt.xlabel('Point Count', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('4th Place Points', fontsize=16)
plt.xticks([i for i in range(60,80,2)],fontsize=10)
plt.yticks([i for i in range(0,4)], fontsize=12)

plt.subplot(133)
plt.hist(df['Safety'], edgecolor='k', color='#B3995D', bins=int(df['Safety'].max()-df['Safety'].min()), align='left')
plt.xticks([i for i in range(34,45,2)], fontsize=10)
plt.xlabel('Point Count', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Safety', fontsize=16)
plt.yticks(fontsize=12)

plt.subplots_adjust(top=0.52, bottom=0.08, left=0.09, right=0.95, hspace=0.5,
                    wspace=0.40)
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;So a little bit about what I was looking at here, what the graphs show us and what we can learn from them. I wanted to look at the point totals in the English Premier League (soccer) since 2000. In this league, there are three big cutoff points that I wanted to look at. Obviously getting first in the league is a big point (there is no playoffs, so most points in the leagues wins it). Fourth place is another big milestone because the top 4 teams get to play in the Champions League in the following season, a world cup type tournament that happens every year with the best teams from all the European leagues from the previous year. Lastly is 17th, which is the last team that is safe to stay in the premier league. 18th-20th end up dropping down to the league beneath and they send up 3 clubs to compete in the Premier League. As a comparison, this would be like if the 3/4 worst baseball teams got sent down to the minor leagues and they sent up the 3/4 best. Dropping down is a huge financial hit and failure for many clubs and staying up is important to compete in the best league. So the biggest decision I had to make was how to measure the point thresholds. If a team won the league with 90 points and 2nd had 80, should I be marking down the 90 points that were earned or set it at 81 since this would have won the league too. I decided to mark it down with what the teams finished with, although I would like to look through it again to set it where it is 1 point above 2nd place, 5th place and 18th place and see how it looks. I should probably also explain that a team earns 3 points for a win, 1 point for a draw and 0 for a loss, and each team places every other team once home and once away during a season. So 38 games (19 other teams played twice), and the total points in the league is not static since draws effect the total number of points in a season.

The only really interesting thing is that twice 80 points was enough to win the league, but once a team got 78 and was only good enough for 4th.

I would say the box plot is the second most helpful, mostly because it is loaded with information. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The most helpful and useful graph was the first one, which was a simple line chart of each of the years. Since each years is different and a team doing well at the top of the league effects the points that other teams are getting, its good to get to see how that effects the different milestones needed to stay safe or to hit the top 4. There are a couple interesting patterns you can see when looking at the graph of each year and the different points. Years that the cutoff for safety is higher, it seems that the points to win the league and to hit the top 4 drop. This is because if the lower ranked teams are getting wins and draws against the better teams, the points are more dispersed and hurt the top teams. Same for the other way around. If the lower teams end up with a low amount of points, the top teams aren't underperforming against them and the league winners and 4th place cutoffs are generally higher. It also shows the low amount of variance in points of being safe (between 34 and 44) and a high variance in the league winners (between 80 and 100). There are also some interesting recent trends in the graphs because the league winner the past three years have ended with 100, 98 and 99 points, 2 of which are outliers and all three are much significantly higher than both the median and mean of the first place points.
